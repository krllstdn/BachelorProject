```python
import pickle
import matplotlib.pyplot as plt
from sksurv.metrics import brier_score
```

# Living

```python
coxnet_pickle_living_file = 'pickle/brier/BS_COXNET_LIVING_FINAL.pickle'
rsf_pickle_living_file = 'pickle/brier/RSF_LIVING_FINAL_BRIER_SCORE.pkl'
gbsa_pickle_living_file = 'pickle/brier/BS_COXNET_LIVING_FINAL.pickle'

with open(coxnet_pickle_living_file, 'rb') as f:
    coxnet_BS = pickle.load(f)

with open(rsf_pickle_living_file, 'rb') as f:
    rsf_BS = pickle.load(f)

with open(gbsa_pickle_living_file, 'rb') as f:
    gbsa_BS = pickle.load(f)
```

```python
coxnet_pickle_living_file = 'pickle/auc/AUC_COXNET_LIVING_FINAL.pickle'
rsf_pickle_living_file = 'pickle/auc/AUC_RSF_LIVING_FINAL.pickle'
gbsa_pickle_living_file = 'pickle/auc/AUC_GBSA_LIVING_FINAL.pickle'

with open(coxnet_pickle_living_file, 'rb') as f:
    cox_time, coxnet_AUC, cox_mean = pickle.load(f)

with open(rsf_pickle_living_file, 'rb') as f:
    rsf_time, rsf_AUC, rsf_mean = pickle.load(f)

with open(gbsa_pickle_living_file, 'rb') as f:
    gbsa_time, gbsa_AUC, gbsa_mean = pickle.load(f)
```

```python
# plt.figure(figsize=(10,6))
plt.plot(coxnet_BS[0], coxnet_BS[1], marker=".", color="orange")
plt.plot(rsf_BS[0], rsf_BS[1], marker=",",)
plt.plot(gbsa_BS[0], gbsa_BS[1], marker=",", color="blue")
# plt.axhline(ibs, linestyle="--")
# plt.text(5, 0, "{model}", fontsize=12)
plt.title("Time-Dependent Brier Score for All Models LDT")
plt.xlabel("days")
plt.ylabel("time-dependent Brier Score")
plt.legend(["Coxnet", "RSF", "GBSA"])
plt.grid(True)
path = "../Thesis/Images/Results/"
plt.savefig(path + "BS_living.pdf", bbox_inches="tight")
plt.show()
```

```python
plt.plot(cox_time, coxnet_AUC, marker=",", color="blue")
plt.plot(rsf_time, rsf_AUC, marker=",", color="orange")
plt.plot(gbsa_time, gbsa_AUC, marker=",", color="green")
plt.axhline(cox_mean, linestyle="--", color="blue")
plt.axhline(rsf_mean, linestyle="--", color="orange")
plt.axhline(gbsa_mean, linestyle="--", color="green")
# plt.text(5, 0, "{model}", fontsize=12)
plt.xlabel("days")
plt.ylabel("time-dependent AUC")
plt.grid(True)
plt.legend(["Coxnet", "RSF", "GBSA", "Coxnet Mean",  "RSF Mean", "GBSA Mean"])
plt.title("Time-Dependent Area Under the ROC for All Models LDT")
path = "../Thesis/Images/Results/"
plt.savefig(path + "AUC_living.pdf", bbox_inches="tight")
plt.show()
```
