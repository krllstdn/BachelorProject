```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_function(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = logistic_function(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, color='blue', label=r'$f(x) = \frac{1}{1 + e^{-x}}$')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.axvline(0, linestyle='--', color='k', label='Decision Boundary')
plt.legend()
plt.grid(True)
plt.title('Logistic Function')
plt.xlim(-10, 10)
path = "../Thesis/Images/"
plt.savefig(path + "logistic_regression.pdf", bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
x = np.linspace(0, 10, 50)
y = 2 * x + 1 + np.random.normal(0, 1, 50)  

coefficients = np.polyfit(x, y, 1)
line = np.poly1d(coefficients)
line_of_best_fit = line(x)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Sample Data')
plt.plot(x, line_of_best_fit, color='blue', label='Line of Best Fit')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')

plt.legend()

plt.grid(True)
plt.xlim(0, 10)
path = "../Thesis/Images/"
plt.savefig(path + "linear_regression.pdf", bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt


patients = ['Patient 1', 'Patient 2', 'Patient 3', 'Patient 4', 'Patient 5']
event_times = [3, 4, 5, 5, 2]  
is_censored = [False, False, True, True, False]  
exit_study_patient = 1  

plt.figure(figsize=(10, 6))

for i, patient in enumerate(patients):
    end_time = event_times[i] if i != exit_study_patient else event_times[i] + 1  
    plt.plot([0, event_times[i]], [i, i], 'k-', lw=1.7) 
    if i == exit_study_patient:
        plt.plot(event_times[i], i, 'bo', markersize=10, label='Exit from study' if i == 1 else "")
    elif is_censored[i]:
        plt.plot(end_time, i, 'k-', lw=2)
    else:
        plt.plot(event_times[i], i, 'rx', markersize=10, label='Event' if i == 0 else "")

plt.axvline(5, color='gray', linestyle='--', lw=2, label='The end of the study')

plt.yticks(range(len(patients)), patients)
plt.xlabel('Time')

plt.title('Censoring in Survival Analysis')
plt.legend(loc='upper right')

path = "../Thesis/Images/"
plt.savefig(path + "censoring_ill.pdf", bbox_inches="tight")

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

scale = 100

x = np.linspace(0, 500, 1000)

survival_function = expon.sf(x, scale=scale)

plt.figure(figsize=(8, 6))
plt.plot(x, survival_function, label=f'Survival Function', color='darkblue')
plt.title('Sample Survival Function')
plt.xlabel('Time')
plt.ylabel('Survival Probability')
plt.xlim(0, 500)
plt.ylim(0, 1)
plt.grid(True)
plt.legend()
path = "../Thesis/Images/"
plt.savefig(path + "survival_f.pdf", bbox_inches="tight")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Time axis
t = np.linspace(0, 10, 100)

# Functions for different hazard types
def decreasing_hazard(t):
    return 1 / (1 + t)

def increasing_hazard(t):
    return t / (1 + t)

def unimodal_hazard(t):
    return t * np.exp(-t)

def constant_hazard(t):
    return np.ones_like(t)


# Plotting
plt.figure(figsize=(10, 6))

plt.plot(t, decreasing_hazard(t), label='Decreasing Hazard', color='darkblue')
plt.plot(t, increasing_hazard(t), label='Increasing Hazard', color="#46b9e0")
plt.plot(t, unimodal_hazard(t), label='Unimodal Hazard')
plt.plot(t, constant_hazard(t), label='Constant Hazard', color='blue')

plt.xlabel('Time')
plt.ylabel('Hazard Rate')

plt.title('Hazard Function Examples')
plt.legend()
plt.grid(True)
plt.xlim(0, 8)
plt.ylim(0, 1.5)

path = "../Thesis/Images/"
plt.savefig(path + "hazard_f.pdf", bbox_inches="tight")

plt.show()
```
