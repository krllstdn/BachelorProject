```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_function(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = logistic_function(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, color='blue', label=r'$f(x) = \frac{1}{1 + e^{-x}}$')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.axvline(0, linestyle='--', color='k', label='Decision Boundary')
plt.legend()
plt.grid(True)
plt.title('Logistic Function')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
x = np.linspace(0, 10, 50)
y = 2 * x + 1 + np.random.normal(0, 1, 50)  

coefficients = np.polyfit(x, y, 1)
line = np.poly1d(coefficients)
line_of_best_fit = line(x)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Sample Data')
plt.plot(x, line_of_best_fit, color='blue', label='Line of Best Fit')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')

plt.legend()

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
red_class = np.random.rand(20, 2) * 2 - 3
blue_class = np.random.rand(20, 2) * 2.3 + 1

x_decision_boundary = np.linspace(-5, 5, 100)
y_decision_boundary = -0.4555 * x_decision_boundary 

plt.figure(figsize=(8, 6))
plt.scatter(red_class[:, 0], red_class[:, 1], color='red',)
plt.scatter(blue_class[:, 0], blue_class[:, 1], color='blue',)

plt.plot(x_decision_boundary, y_decision_boundary, color='black', label='Decision Boundary')
plt.plot(x_decision_boundary, y_decision_boundary + 1.8, 'k--', label='Support Vectors')
plt.plot(x_decision_boundary, y_decision_boundary - 1.8, 'k--')

plt.scatter([-1, 1.3], [-1.35, 1.2], color='grey', marker='x', s=100, label='Support Vectors')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM Demonstration for Two-Dimensional Dataset')

plt.legend(loc='upper left')

plt.text(2.8, -1.5, r'$\vec{w}\cdot\vec{x} - b = 0$', fontsize=12, ha='center', va='bottom', rotation=-23)

plt.text(2.8, 0.3, r'$\vec{w}\cdot\vec{x} - b = 1$', fontsize=12, ha='center', va='bottom', rotation=-23)
plt.text(2.8, -3.35, r'$\vec{w}\cdot\vec{x} - b = -1$', fontsize=12, ha='center', va='bottom', rotation=-23)

plt.arrow(-3, -.36, 1.5, 2.75, head_width=0.1, head_length=0.1, fc='r', ec='r', lw=2, length_includes_head=True,)
plt.arrow(-1.5, 2.4, -1.5, -2.75, head_width=0.1, head_length=0.1, fc='r', ec='r', lw=2, length_includes_head=True,)

plt.text(-2.8, 0.3, r'$\frac{2}{||\overline{w}||}$', fontsize=14, ha='center', va='bottom', rotation=60)

plt.grid(True)
# plt.axis('equal')
plt.ylim(-3.3, 3.3)
plt.xlim(-4, 4)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for demonstration
np.random.seed(0)
red_class = np.random.rand(20, 2) * 4 - 2  # 20 points for the red class
blue_class = np.random.rand(20, 2) * 4 - 2  # 20 points for the blue class

# Add a cluster of blue points in the middle of the red points
blue_class = np.vstack((blue_class, np.array([[0, 0]])))

# Plot the synthetic data
plt.figure(figsize=(8, 6))
plt.scatter(red_class[:, 0], red_class[:, 1], color='red', label='Red Class')
plt.scatter(blue_class[:, 0], blue_class[:, 1], color='blue', label='Blue Class')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linearly Inseparable Dataset')

# Move the legend to the top left corner
# plt.legend(loc='upper left')

# Show the plot
plt.grid(True)
plt.show()
```
